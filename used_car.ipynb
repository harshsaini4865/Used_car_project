```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\om sai compurer1\\Desktop\\used car\\CAR DETAILS FROM CAR DEKHO.csv")

```

```python
df
```

```python
df["fuel"].unique()
```

```python
fuel ={
    "Petrol": 0,
    "Diesel": 1,
    "CNG": 2,
    "LPG": 3,
    "Electric": 4,
}
```

```python
df["fuel"]=df["fuel"].map(fuel)
```

```python
df["seller_type"].unique()
```

```python
seller_type = {
    "Individual": 0,
    "Dealer": 1,
    "Trustmark Dealer": 2
}
```

```python
df["seller_type"]=df["seller_type"].map(seller_type)
```

```python
df["transmission"].unique()
transmission = {
    "Manual": 0,
    "Automatic": 1,
}
df["transmission"]=df["transmission"].map(transmission)
df
```

```python
df["owner"].unique()
owner = {
    "First Owner": 0,
    "Second Owner": 1,
    "Third Owner": 2,
    "Fourth & Above Owner": 3,
    "Test Drive Car": 4,
}
df["owner"]=df["owner"].map(owner)
df
```

```python
x=df.drop(columns=["selling_price","name"])
y=df["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.predict(x_test)
```

```python
lr.predict(x_train)
```
